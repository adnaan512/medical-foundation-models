# HAM10000 — Exploratory Data Analysis

Dataset statistics, class distribution, sample images, and augmentation preview.

```python
import sys
sys.path.append('..')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
from datasets import HAM10000Dataset, CLASS_NAMES, IDX_TO_CLASS
from datasets.transforms import get_val_transforms, get_train_transforms, denormalise
DATA_PATH = '../data/HAM10000'
SEED = 42
```

```python
transform = get_val_transforms(224)
train_ds = HAM10000Dataset(DATA_PATH, 'train', transform=transform, seed=SEED)
val_ds   = HAM10000Dataset(DATA_PATH, 'val',   transform=transform, seed=SEED)
test_ds  = HAM10000Dataset(DATA_PATH, 'test',  transform=transform, seed=SEED)
from datasets import print_dataset_summary
print_dataset_summary(train_ds, val_ds, test_ds)
```

```python
dist = train_ds.get_class_distribution()
fig, ax = plt.subplots(figsize=(10,5))
colors=['#2196F3','#F44336','#4CAF50','#FF9800','#9C27B0','#00BCD4','#795548']
bars=ax.bar(list(dist.keys()),list(dist.values()),color=colors,edgecolor='black')
ax.set_title('HAM10000 Training Set Class Distribution',fontweight='bold')
for bar in bars:
    h=bar.get_height()
    ax.text(bar.get_x()+bar.get_width()/2.,h+20,f'{h:,}',ha='center',fontsize=9)
plt.tight_layout()
plt.savefig('../figures/class_distribution.png',dpi=150)
plt.show()
```

```python
weights=train_ds.class_weights
print(pd.DataFrame({'class':list(IDX_TO_CLASS.values()),'weight':weights}).sort_values('weight',ascending=False).to_string(index=False))
```
